```python
df=df.drop(['VeilType'],axis=1)
df=df.drop(['StalkRoot'],axis=1)
```

```python
from sklearn import preprocessing

lab_enc=preprocessing.LabelEncoder()
```

```python
coloumns1=df.columns
```

```python
for index in coloumns1:
    df[index]=pd.Series(lab_enc.fit_transform(df[index]))
```

```python
X=df.drop(["Class"],axis=1)
```

```python
Y=df["Class"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x,x_test,y,y_test=train_test_split(X,Y,test_size=.2,random_state=42)
```

```python
x_train,x_val,y_train,y_val=train_test_split(x,y,test_size=.25,random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import BernoulliNB
```

```python
param_tree={'criterion':['gini', 'entropy'], 'splitter':['best','random'],'max_depth':[4,5,6,8,7,9]}

param_forest={'criterion':['gini', 'entropy'],'max_depth':[4,5,6,8,7,9], 'n_estimators':[40,60,80,100] }

param_naive={"alpha":[1,2,3]}
```

```python
grid_tree=GridSearchCV(DecisionTreeClassifier(),param_tree,cv=7)
grid_forest=GridSearchCV(RandomForestClassifier(),param_forest,cv=7)
grid_naive=GridSearchCV(BernoulliNB(),param_naive,cv=7)
```

```python
grid_tree.fit(x_train,y_train)
grid_forest.fit(x_train,y_train)
grid_naive.fit(x_train,y_train)
```

```python
y_pred1=grid_naive.predict(x_val)
y_pred2=grid_tree.predict(x_val)
y_pred3=grid_forest.predict(x_val)
```

```python
from sklearn.metrics import accuracy_score
print(f"Accuracy by naive baise classifier : {accuracy_score(y_val,y_pred1)}")
print(f"Accuracy by Decision tree classifier : {accuracy_score(y_val,y_pred2)}")
print(f"Accuracy by Random Forest classifier : {accuracy_score(y_val,y_pred3)}")


```
